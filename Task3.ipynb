## Visualize data

### Subtask:
Create visualizations to represent the key trends and insights from the data.


**Reasoning**:
I need to create several visualizations to represent the key trends and insights identified in the analysis phase. I will group the bar plots for categorical variables in one code block and the box/violin plots for numerical variables in another code block to keep the code organized. I will start with the bar plots for categorical variables as requested in steps 1-8 of the instructions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_features = ['job', 'marital', 'education', 'housing', 'loan', 'contact', 'month', 'poutcome']

plt.figure(figsize=(20, 15))
for i, feature in enumerate(categorical_features):
    plt.subplot(3, 3, i + 1)
    sns.countplot(data=df, x=feature, hue='deposit', palette='viridis')
    plt.title(f'Deposit Proportion by {feature.capitalize()}')
    plt.xlabel(feature.capitalize())
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

plt.show()
```

**Reasoning**:
I have created bar plots for the categorical features. Now, as per steps 9-12, I need to create box plots or violin plots to visualize the distribution of numerical features ('duration', 'campaign', 'pdays', and 'previous') for 'yes' and 'no' deposits. I will use box plots and handle the '-1' values in 'pdays' as instructed.



```python
numerical_features = ['duration', 'campaign', 'pdays', 'previous']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 2, i + 1)
    if feature == 'pdays':
        # Filter out -1 values for pdays
        sns.boxplot(data=df[df['pdays'] != -1], x='deposit', y=feature, palette='viridis')
    else:
        sns.boxplot(data=df, x='deposit', y=feature, palette='viridis')
    plt.title(f'Distribution of {feature.capitalize()} by Deposit')
    plt.xlabel('Deposit')
    plt.ylabel(feature.capitalize())
    plt.tight_layout()

plt.show()
```

## Create dashboard

### Subtask:
Combine the visualizations into a dashboard for easy interpretation and presentation.


**Reasoning**:
Combine the previously generated count plots and box plots into a single dashboard using subplots to facilitate easy interpretation and presentation.



**Reasoning**:
The previous command failed because 'ha' is not a valid keyword argument for `tick_params`. The horizontal alignment of tick labels should be set using `set_xticklabels` after setting the rotation. The code needs to be corrected to use `set_xticklabels` with the appropriate alignment and rotation.



```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_features = ['job', 'marital', 'education', 'housing', 'loan', 'contact', 'month', 'poutcome']
numerical_features = ['duration', 'campaign', 'pdays', 'previous']

fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(25, 18))
fig.suptitle("Bank Marketing Campaign Analysis Dashboard", fontsize=20)

# Plotting count plots for categorical features
for i, feature in enumerate(categorical_features):
    row = i // 4
    col = i % 4
    sns.countplot(data=df, x=feature, hue='deposit', palette='viridis', ax=axes[row, col])
    axes[row, col].set_title(f'Deposit Proportion by {feature.capitalize()}')
    axes[row, col].set_xlabel(feature.capitalize())
    axes[row, col].set_ylabel('Count')
    axes[row, col].tick_params(axis='x', rotation=45)
    axes[row, col].set_xticklabels(axes[row, col].get_xticklabels(), rotation=45, ha='right')
    if i < len(categorical_features) - 1:  # Hide legend for all but the last count plot
        axes[row, col].legend_.remove()


# Plotting box plots for numerical features
for i, feature in enumerate(numerical_features):
    row = (len(categorical_features) + i) // 4
    col = (len(categorical_features) + i) % 4
    if feature == 'pdays':
        sns.boxplot(data=df[df['pdays'] != -1], x='deposit', y=feature, palette='viridis', ax=axes[row, col])
    else:
        sns.boxplot(data=df, x='deposit', y=feature, palette='viridis', ax=axes[row, col])
    axes[row, col].set_title(f'Distribution of {feature.capitalize()} by Deposit')
    axes[row, col].set_xlabel('Deposit')
    axes[row, col].set_ylabel(feature.capitalize())

# Remove any unused subplots
for i in range(len(categorical_features) + len(numerical_features), 12):
    row = i // 4
    col = i % 4
    fig.delaxes(axes[row, col])


plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
plt.show()
```
